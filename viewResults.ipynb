```python
import h5py
import numpy as np
import matplotlib.pyplot as plt


def compare_plasma_files(file1, file2):

    species = ["He 1+", "e-"]

    with h5py.File(file1, "r") as f1, h5py.File(file2, "r") as f2:

        for sp in species:

            # =========================
            # Instant density
            # =========================
            dens1 = np.array(f1[sp]["instant density"])
            dens2 = np.array(f2[sp]["instant density"])

            plt.figure()
            plt.plot(dens1, label=f"before merging - {sp}")
            plt.plot(dens2, label=f"after merging - {sp}")
            plt.title(f"{sp} - Instant Density")
            plt.xlabel("cell index")
            plt.ylabel("Density (m^-3)")
            plt.legend()
            plt.grid()
            plt.show()

            # =========================
            # Velocity distribution function
            # =========================
            vdf1 = np.array(f1[sp]["instant_vdf"])
            bins1 = np.array(f1[sp]["instant_vdf_bins"])

            vdf2 = np.array(f2[sp]["instant_vdf"])
            bins2 = np.array(f2[sp]["instant_vdf_bins"])

            plt.figure()
            plt.plot(bins1, vdf1, label=f"before_merging - {sp}")
            plt.plot(bins2, vdf2, label=f"after_merging - {sp}")
            plt.title(f"{sp} - Instant VDF")
            plt.xlabel("Velocity (m/s)")
            plt.ylabel("nombre de particules par unité de vitesse par unité de longueur spatiale")
            plt.legend()
            plt.grid()
            plt.show()
```
